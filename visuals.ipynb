```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Create the synthetic data
data = pd.DataFrame({
    "region": ["West", "West", "East", "East", "South", "South", "West", "East"],
    "product": ["A", "B", "A", "B", "A", "B", "A", "A"],
    "revenue": [200, 150, 300, 120, 180, 220, 250, 275],
    "quantity": [2, 1, 3, 1, 2, 2, 2, 3]
})

data
```

```python
# Creating a Bar Chart(Comparing catergories)

# What is the total
# revenue by region


# Groupby Aggregation
revenue_by_region = (
    data
    .groupby("region")["revenue"]
    .sum()
    .reset_index()
)

revenue_by_region
```

```python
# Creating the bar chart
plt.figure()
plt.bar(revenue_by_region["region"], revenue_by_region["revenue"])
plt.xlabel("Region")
plt.ylabel("Total Revenue")
plt.title("Total Revenue by Region")
plt.show()
```

```python
# Line Chart(Trends over time)

time_data = pd.DataFrame({
    "date" : pd.date_range("2025-01-01", periods = 8, freq = "D"),
    "revenue" : [200, 220, 210, 240, 260, 255, 270, 280]
})

plt.figure()
plt.plot(time_data["date"], time_data["revenue"], marker = "o")
plt.xlabel("Date")
plt.ylabel("Revenue")
plt.title("Daily Revenue Trend")
plt.show()
```

```python
# Creating a Histogram(Distributions)

# How is revenue distributed

plt.figure()
plt.hist(data["revenue"], bins=5)
plt.xlabel("Revenue")
plt.ylabel("Frequency")
plt.title("Distribution of Revenue")
plt.show()
```

```python
# Creating a box plot(Compare distributions, medians, and outliers)

# How does revenue vary by region

plt.figure()
data.boxplot(column="revenue", by="region")
plt.xlabel("Region")
plt.ylabel("Revenue")
plt.title("Revenue Distribution by Region")
plt.suptitle("")  # removes automatic subtitle
plt.show()
```

```python
# More practice

# Create a Bar Chart that shows avg revenue by product

avg_revenue_prod = (
    data
    .groupby("product")["revenue"]
    .mean()
    .reset_index()
)

plt.figure()
plt.bar(avg_revenue_prod["product"], avg_revenue_prod["revenue"])
plt.xlabel("Product")
plt.ylabel("Average Revenue")
plt.title("Average Revenue per product")
plt.show()
```

```python
# Create a histogram for quantity

plt.figure()
plt.hist(data["quantity"], bins = 3)
plt.xlabel("Quantity")
plt.ylabel("Frequency")
plt.title("Distribution of Quantity")
plt.show()
```
